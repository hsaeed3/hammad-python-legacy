```python
from hammadpy import Core

core = Core()
```

```python
"""Text Styles"""
# core.say (
    # message: the message to be printed
    # color: the color of the message | Optional
    # bg: the background color of the message | Optional
    # style: the style of the message | Optional
# )

core.say(message="Hello!", color="yellow", bg="black", style="bold")
core.say("Hello!", "black", "white", "bold")
core.say("Hello!", "cyan")
core.say("Hello!", bg="red")
```
